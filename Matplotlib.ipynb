## MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

Plotting using matplotlib

Plot Scatter

```python
plt.scatter(x,y,c='g')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title(" Scatter plot")
plt.savefig("F:\Data Science\Git hub projects\Practice\scatter.png")
```

```python
y=x*x
```

plt plot

```python
plt.plot(x,y,"r*",linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("2D diagrom")
```

creating subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y*')
```

Compute the x and y coordinates for points on a sine curve 

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("Sine wave form")
plt.plot(x,y)
```

subplot()

Compute the x and y coordinates for points on sine and cosine curves 

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

Set up a subplot grid that has height 2 and width 1, 
and set the first such subplot as active. 

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g-')
plt.title('Cosine')
```

# Bar plot

```python
x=[3,6,8]
y=[9,12,5]
x2=[4,7,9]
y2=[1,9,3]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Bar plot")
```

# Box plot using matplotlib

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False)
```

# Pie chart

```python
labels="Python","C++","Java","Ruby"
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0) #explode 1st slice
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=False)
plt.axis('equal')
```
